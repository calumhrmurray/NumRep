```python
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class gaussian:
    
    # a*math.exp(-(x-mean)/(2*c*c))
    def __init__(self, mean, width):
        self.mean = mean
        self.width = width
        
    def evaluate_gaussian(self,x):
        return math.exp(-(x-self.mean)**2./(2.*self.width**2.))
    
    def maximum(self):
        return 1.0
        
    def random_val(self):        
        x1 = 0 
        y1 = 0
        y2 = 1
        
        while ( y2 > y1 ):
            x1 = (np.random.uniform() -0.5) * 2.0
            x1 = x1 * self.width * 5.0 + self.mean
            y1 = self.evaluate_gaussian( x1)
            y2 = np.random.uniform() * self.maximum()

        return x1
    
    def numeric_integral(self,lo,hi,npoints=1000000):
        
        sumf = 0
        
        for i in range(0,npoints):
            x = lo+np.random.uniform()*(hi-lo)
            sumf += self.evaluate_gaussian(x)
            
        return sumf*(hi-lo)/npoints
    
    def analytic_integral(self):
        return self.width*math.sqrt(2*math.pi)        
    
```

```python
a = gaussian(0,5)
```

```python
plt.figure(figsize=(20,10))
plt.grid(True)
plt.xlim(-15,15)
plt.plot([x for x in range(-100,100)],[a.evaluate_gaussian(x) for x in range(-100,100)],'-')
```

```python
results = []
for i in range(0,100000):
    results.append(a.random_val())    
```

```python
n, bins, patches = plt.hist(results,bins=30)
plt.close()

plt.figure(figsize=(20,10))
plt.plot(bins[1:],n)
plt.grid(True)
```
